# Model Building 

```python
data = customer.pivot_table(index="product", columns="email", values="rating").fillna(0)
data.head()
```

```python
from scipy.sparse import csr_matrix

data_matrix = csr_matrix(data.values) # converting pivot table to array matrix

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(data_matrix)
```

```python
data.shape
```

```python
index = np.random.choice(data.shape[0])
print(f"cusotmer index: {index}\ncustomer mail: {customer.iloc[index,1]}") # random customer
distances, indices = model_knn.kneighbors(data.iloc[index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print(f"Recommendations for {data.index[index]} bank product: \n")
    else:
        print(f"{data.index[indices.flatten()[i]]}, with distance of {distances.flatten()[i]}")
```
